```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import statistics
```

```python
#Loading Dataset
data = pd.read_csv("Google_Stock_Price_Train.csv")
```

```python
#View First 5 rows of Dataset
data.head()
```

```python
#View Last 5 rows of Dataset
data.tail()
```

```python
data.describe()
```

```python
data.info()
```

```python
data["Close"]=data["Close"].str.replace(',','')
data["Close"]=data["Close"].astype(float)
```

```python
data["Volume"]=data["Volume"].str.replace(',','')
data["Volume"]=data["Volume"].astype(float)
```

```python
data["Volume"]=data["Volume"].astype(int)
```

```python
data["Date"] = pd.to_datetime(data["Date"])
```

```python
data.isna().sum()
```

```python
sns.boxplot(data=data[["Open", "High", "Low", "Close"]])
```

```python
data['year'] = data['Date'].dt.year
data.head(5)
```

```python
sns.boxplot(x='year', y='Close', data=data)
plt.title('Closing Stock Prices by Year')
plt.xlabel('Year')
plt.ylabel('Closing Stock Price')
plt.show()

#These plots show that there is something wrong with 2013 and 2014 cloaing stock prices
```

```python
g = data.groupby(data['year'])
```

```python
g[["Open","Close"]].describe()
```

```python
#Stock Closing Price Over Time
plt.figure(figsize=(15, 6))
plt.plot(data['Date'], data['Open'])
plt.xlabel('Date')
plt.ylabel('Opening Price')
plt.title('Stock Opening Price Over Time')
plt.grid(True)
plt.show()
```

```python
#Stock Closing Price Over Time
plt.figure(figsize=(15, 6))
plt.plot(data['Date'], data['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Closing Price Over Time')
plt.grid(True)
plt.show()
```

```python
#Stock Volume Over Time
plt.figure(figsize=(15, 6))
plt.plot(data['Date'], data['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Stock Volume Over Time')
plt.grid(True)
plt.show()
```

```python
#Distribution of Closing Stock Prices
plt.figure(figsize=(10, 6))
sns.histplot(data['Close'], kde=True)
plt.title('Distribution of Closing Stock Price')
plt.xlabel('Closing Stock Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Volume reltion with Closing Price
plt.figure(figsize=(8, 4))
plt.scatter(data['Close'], data['Volume'])
plt.xlabel('Closing Price')
plt.ylabel('Volume')
plt.title('Volume vs. Closing Price')
plt.show()
```

```python
#Scatter Plot between Opening and Closing Price
plt.figure(figsize=(16, 4))
plt.scatter(data['Open'], data['Close'])
plt.xlabel("Opening Price")
plt.ylabel("Closing Price")
plt.title("Opening vs. Closing Price")
plt.show()
```

```python
data["Volatility"]= data['High']-data['Low']
```

```python
#Stock Volatility Over Time, More Volatilty means Stockis not Stable
x= data['High']-data['Low']
plt.figure(figsize=(15, 6))
plt.plot(data['Date'], x)
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.title('Stock Volatility Over Time')
plt.grid(True)
plt.show()
```

```python
#Top 10 Volatility of Stock
data.nlargest(10, 'Volatility')
```

```python
#Total Volume of Stock in each Year
data.groupby("year")["Volume"].sum().plot(kind='bar',figsize=(16,8), color='y' )
plt.title('Total Volume in Each Year', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Total Volume', fontsize=15)
```

```python
#Top 10 Volume with respect to date
top_10_rows = data.nlargest(10, 'Volume')
top_10_rows.plot(x='Date', y=['Volume'], kind='bar', figsize=(16,8), color='purple')
plt.title('Top 10 Volume of Stock Market', fontsize=15)
plt.xlabel('Date', fontsize=15)
plt.xticks(rotation=90, fontsize=12)
plt.ylabel('Volume', fontsize=15)
plt.show()
```

```python
#Highest Volume Each Year
data.groupby('year')['Volume'].max().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(16,8), color='blue')
plt.title('Highest Volume Each Year', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.xticks(rotation=90,fontsize=12)
plt.ylabel('Volume', fontsize=15)
```

```python
#Top 10 Oening of Stocks Overall, This shows that most highest 
top_10_rows = data.nlargest(10, 'Open')
top_10_rows.plot(x='Date', y=['Open'], kind='bar', figsize=(16,8), color='purple')
plt.title('Top 10 Opening of Stock Market', fontsize=15)
plt.xlabel('Date', fontsize=15)
plt.xticks(rotation=90, fontsize=12)
plt.ylabel('Opening Stock', fontsize=15)
plt.show()
```
